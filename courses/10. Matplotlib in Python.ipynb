## Introduce the dataset

In this section, we will try to answer the following questions:

* read 'unrate.csv' in a dataframe and assign this result to the variable unrate
* use the pandas.to_datetime() function to convert the 'DATE' column into datetime values
* display the first 12 values of unrate corresponding to a whole year

```python
import pandas as pd
```

```python
unrate = pd.read_csv("unrate.csv")
unrate.head()
```

## Build graph with our dataset

## Build figure with our dataset

In this section, we will try to answer the following questions:

* create 2 linear curves in a 2 row and 1 column grid:
    * in the top subgraph, plot the 1948 data:
        * take for the x axis the first 12 values of the 'DATE' column
        * take for the y-axis the first 12 values of the 'UNRATE' column
    * in the bottom sub-graph, plot the 1949 data:
        * take for the x-axis the index values 12 to 24 of the 'DATE' column (the next 12 values of the dataframe)
        * take for the y-axis the index values 12 to 24 of the 'UNRATE' column
    * use plt.show() to display the 2 graphs

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['UNRATE'])
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['UNRATE'])

plt.show()
```

## Improve display

In this section, we will try to answer the following questions:

* for the previous graphic, add a width of 15 and a height of 8
* and add a title for the 2 graphs:
    * for ax1: "Unemployment rate in 1948"
    * for ax2: "Unemployment rate in 1949"

```python
fig = plt.figure(figsize=(15, 8))

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['UNRATE'])
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['UNRATE'])

ax1.set_title("Unemployment rate in 1948")
ax2.set_title("Unemployment rate in 1949")

plt.show()
```

## Graph comparison

```python
fig = plt.figure(figsize=(12, 12))

for i in range(5):
    
    ax = fig.add_subplot(5, 1, i+1)
    
    start_index = i*12
    end_index = (i+1)*12
    
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['UNRATE'])
    
plt.show()
```

## Curve superposition

In this section, we will try to answer the following questions:

* set the display area to 6 inches wide and 3 inches high
* generate 2 curves in one graph by using the 'MONTH' column for the x axis instead of the 'DATE' column:
    * one curve with the 1948 values with 'red' color
    * a curve with 1949 values in 'blue' color
* use plt.show() to display the curves

```python
unrate['MONTH'] = unrate['DATE'].dt.month

plt.figure(figsize=(6, 3))

plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['UNRATE'], c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['UNRATE'], c='blue')
                                                 
plt.show()
```

## Add more curves

In this section, we will try to answer the following questions:

* set the graph display area to 15 inches wide and 6 inches high
* generate the following curves on the same graph:
    * 1948: red curve
    * 1949: blue curve
    * 1950: green curve
    * 1951: orange curve 'orange' 
    * 1952: black curve 'black
* use plt.show() to display the curves

```python
plt.figure(figsize=(15, 6))

colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
    
    start_index = i*12
    end_index = (i+1)*12
    
    subset = unrate[start_index:end_index]
    
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i])
    
plt.show()
```

## Add a legend

In this section, we will try to answer the following questions:

* modify the previous code to include a legend specifying the year for each color:
    * i.e. the 1948 data curve that uses the color red will have the caption '1948' in the legend
* position the legend in the upper left corner
* display the graph

```python
plt.figure(figsize=(15, 6))

colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
    
    start_index = i*12
    end_index = (i+1)*12
    
    subset = unrate[start_index:end_index]
    
    labels = str(1948+i)
    
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=labels)

plt.legend(loc='upper left')
plt.show()
```

## Final curve

In this section, we will try to answer the following questions:

* modify the previous code to:
    * add the title 'Evolution of the unemployment rate from 1948 to 1952'
    * add the title for the x axis 'Months (from 1 to 12)'
    * add the heading for the y-axis 'Unemployment rate (in %)'.

```python
plt.figure(figsize=(15, 6))

colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
    
    start_index = i*12
    end_index = (i+1)*12
    
    subset = unrate[start_index:end_index]
    
    labels = str(1948+i)
    
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=labels)

plt.legend(loc='upper left')
plt.xlabel("Months (from 1 to 12)")
plt.ylabel("Unemployment rate (in %)")
plt.title("Evolution of the unemployment rate from 1948 to 1952")
plt.show()
```
