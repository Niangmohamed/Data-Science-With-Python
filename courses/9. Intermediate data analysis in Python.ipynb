## Convert age to a numeric value

In this section, we will try to answer the following questions:

* write a function that converts a string into an integer value. This will convert the values in the the 'Age' column into integers. This function will take as parameter a string of characters (the current values of the column 'Age'):
    * use the isnull() method to check if the values are null. Add an if condition that returns 'None' if the value is null
    * split the strings according to the space (' ') and extract the first element of the resulting list (split() method)
    * replace the '+' character in the result with an empty string to remove it (replace() method)
    * use the int() function to convert the result into an integer
    * return the result
* use the series.apply() method to apply the function to each value of the 'Age' column of the dataframe df
* assign the result to the new age_int column of the dataframe
* call the series.describe() method on the 'age_int' column of the dataframe df and display the result

```python
df["Age"].value_counts()
```

```python
def extract_age(age_str):
    
    if pd.isnull(age_str):
        return None
    
    age_str = age_str.split(" ")[0]
    age_str = age_str.replace("+", "")
    age_int = int(age_str)
    
    return age_int
```

```python
df["age_int"] = df["Age"].apply(extract_age)
df["age_int"].describe()
```

## Convert income into numerical values

In this section, we will try to answer the following questions:

* write a function to convert the income to a single integer value:
    * use the isnull() method to check if the value is null. If it is, return 'None'
    * separate the string by taking the space as delimiter and extract the first element of the resulting list
    * if the result is 'Prefer', return 'none
    * replace the characters '$' and ',' with empty strings to remove them
    * use the int() function to convert the result to an integer
    * return the result
* use the series.apply() method to apply the previous function to each value in the 'How much total combined money did all members of your HOUSEHOLD earn last year?':
    * assign the result to the new column 'income_int' of the dataframe df
* call the series.describe() method on the income_int column of the df dataframe and display the result

```python
df["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts()
```

```python
def extract_income(income_str):

    if pd.isnull(income_str):
        return None

    income_str = income_str.split(" ")[0]
    if income_str == "Prefer":
        return None

    income_str = income_str.replace(",", "")
    income_str = income_str.replace("$", "")

    income_int = int(income_str)

    return income_int
```

```python
df["income_int"] = df["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(extract_income)
df["income_int"].describe()
```
