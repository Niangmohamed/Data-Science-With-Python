## Use custom indexes

```python
fandango_ind_film = fandango.set_index(keys="FILM", drop=False)
fandango_ind_film.head()
```

```python
fandango_ind_film.index
```

## Selecting values from a custom index

```python
fandango_ind_film['Danny Collins (2015)':'Serena (2015)']
```

```python
fandango_ind_film.loc['Danny Collins (2015)':'Serena (2015)']
```

```python
movies = ["Danny Collins (2015)", "Spare Parts (2015)", "Serena (2015)"]
fandango_ind_film.loc[movies]
```

## Apply() method on the columns of a dataframe

```python
import numpy as np
```

```python
types = fandango_ind_film.dtypes
types
```

```python
float_columns = types[types.values == 'float64'].index
float_columns
```

```python
float_fandando_ind_film = fandango_ind_film[float_columns]
float_fandando_ind_film.head()
```

```python
std = float_fandando_ind_film.apply(lambda x: np.std(x))
std
```

```python
by_two = float_fandando_ind_film.apply(lambda x: x/2)
by_two.head()
```

## Apply() method on the rows of a dataframe

```python
rt_mt_user = float_fandando_ind_film[['RT_user_norm', 'Metacritic_user_nom']] 
rt_mt_user
```

```python
rt_mt_user.apply(lambda x: np.std(x), axis=1)
```

```python
rt_mt_means = rt_mt_user.apply(lambda x: np.mean(x), axis=1)
rt_mt_means
```
